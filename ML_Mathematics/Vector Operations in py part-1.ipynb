```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

**Plotting the vector**

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy',angles='xy',scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy',angles='xy',scale=1,color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,-3,-6, scale_units='xy',angles='xy',scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

**Addition of Vectors**

```python
vector1=np.array([0,0,2,3])
vector2=np.array([0,0,3,-2])

sum = vector1+vector2
print(sum)
```

```python
plt.quiver(0,0,2,3, scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,3,-2, scale_units='xy',angles='xy',scale=1)
plt.quiver(0,0,5,1, scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

**Subtraction of two vectors**

```python
vector1=np.array([0,0,2,3])
vector2=np.array([0,0,3,-2])

sub = vector1-vector2
print(sub)
```

```python
plt.quiver(0,0,2,3, scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,3,-2, scale_units='xy',angles='xy',scale=1)
plt.quiver(0,0,-1,5, scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

**Multiplication of two vectors**

```python
vector1=np.array([0,0,2,3])
vector2=np.array([0,0,3,-2])

mul = vector1*vector2
print(mul)
```

```python
plt.quiver(0,0,2,3, scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,3,-2, scale_units='xy',angles='xy',scale=1)
plt.quiver(0,0,6,-6, scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

**or**

```python
vector1=np.array([0,0,2,3])
vector2=2*vector1
print(vector2)
```
