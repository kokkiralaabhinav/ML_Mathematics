```python
import numpy as np
```

**Creating a Matrix using Numpy**

```python
m1 = np.array([[2,3],[6,7]])
print(m1)
```

```python
m2 = np.array([[2,3],[6,7]])
print(m2)
```

```python
m1.shape
m2.shape
```

```python
print(m1+m2)
```

```python
m3 = np.array([[10,35,45],[50,64,80],[20,15,90]])
print(m3)
```

**Creating Matrices with random values**

```python
rm = np.random.rand(3,3)
print(rm)
```

```python
# Creating matrix with random integers
rmi = np.random.randint(100, size=(3,3))
print(rmi)
```

**Matrix with all values as one**

```python
m4 = np.ones((2,3))
print(m4)
```

```python
m5 = np.ones((2,3), dtype=int)
print(m5)
```

```python
m5 = np.ones((4,4), dtype=int)
print(m5)
```

**Matrix with all values as Zeros:null matrix or zero matrix**

```python
m6 = np.zeros((4,4))
print(m6)
```

```python
m7 = np.zeros((4,4), dtype=int)
print(m7)
```

**Identity matrix**

```python
m8 = np.eye(4,4)
print(m8)
```

**Transpose of a Matrix**

```python
# Matrix with random integers values
a = np.random.randint(100,size=(4,5))
print(a)
```

```python
transpose = np.transpose(a)
print(transpose)
```
